```python
def parse_category_rambler(category, pages, output):
    SCROLL_PAUSE_TIME = 1
    prefs = {"profile.managed_default_content_settings.images": 2,
             "profile.default_content_settings.cookies": 2}
    chrome_options = Options()
    chrome_options.add_extension('adblock.crx')
    chrome_options.add_experimental_option("prefs",prefs)
    chrome_options.add_argument('--disable-application-cache')
    driver = Chrome(executable_path=chrome_driver_path, chrome_options=chrome_options)
    driver.implicitly_wait(10)
#     driver.close()
    print("Current session is {}".format(driver.session_id))
    print(driver.window_handles)
    driver.switch_to.window(driver.window_handles[0])
    driver.get('chrome-extension://gighmmpiobklfepjocnamgkkbiglidom/options.html')
    driver.find_element_by_id('acceptable_ads').click()#adblock customization
#     driver.get('https://news.rambler.ru/politics/latest/?page=1')
#     driver.get('https://life.ru/c/politika/')
#     driver.find_element_by_class_name('j-footer__switch').click()
#     driver.find_element_by_class_name('list-item__title').cli
    
    driver.get('https://life.ru/' + category)
    for page in tqdm.tqdm(range(0, pages),position=1):
        #for element in tqdm.tqdm(driver.find_elements_by_class_name('article-card'),position=0):
#         for element in tqdm.tqdm(driver.find_elements_by_xpath("//article"),position=0):


        for element in tqdm.tqdm(driver.find_elements_by_class_name("styles_root__2aHN8")[page*30:],position=0):
#             print(element.get_attribute('href'))
#             print(element.find_element_by_tag_name('p').text)
            try:
                with open(output,'a', encoding='utf-8') as file:
                    file.write(json.dumps({
                        'link' : element.get_attribute('href'),
                        'title' : element.find_element_by_tag_name('p').text,
                        'category': category[2:]
                    },ensure_ascii=False) + '\n')
            except:
                continue
        
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        # Wait to load page
        time.sleep(SCROLL_PAUSE_TIME)
    driver.quit()
```

# 2. Парсинг Rambler новостей (BeautifulSoup + html2text)

```python
posts = []
with open(links_filename, encoding='utf-8') as file:
    for line in file:
        posts.append(json.loads(line))
```

```python
print(len(posts))
```

```python
posts[0]
```

```python
id_to_del = []
for i in range(len(posts)):
    for j in range(i, len(posts)):
        if i != j and posts[i]['title'] == posts[j]['title']:
            id_to_del.append(posts[j])
print(len(id_to_del))
```

```python
id_to_del[0]
```

```python
for i in range(len(id_to_del)):
    for j in range(len(posts) - len(id_to_del)):
        if posts[j]['title'] == id_to_del[i]['title']:
            posts.pop(j)
            break
print(len(posts))
```

```python
len(posts) + len(id_to_del)
```

```python
for element in tqdm.tqdm(posts[876:]):
#     print(element['link'])
    soup = BeautifulSoup(requests.get(element['link']).text, "lxml")
    text = ''
    tags = []
    for paragraph in soup.findAll('div', attrs={'class': 'indentRules_block__iwiZV styles_text__3IVkI'}):
        for child in paragraph.findAll('p', attrs={'class': ''}):
            text += html2text(child.text).replace('\n', ' ') + ' '
#     print('==========={}==========={}'.format(element['title'], element['link']))
#     print(text)
    for tag in soup.findAll('a', attrs={'class': 'styles_tag__1D3vf'}):
        tags.append(tag.text)

    with open('rambler_news_tags.json','a') as file:
        file.write(json.dumps({
            'link' : element['link'],
            'title' : element['title'],
            'category': element['category'],
            'tags' : tags,
            'text': text,
        },ensure_ascii=True) + '\n')
```
